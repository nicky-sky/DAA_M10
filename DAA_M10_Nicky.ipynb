```python
from scipy.optimize import linprog

c = [0, 0]
A_eq = [
    [4, 12],
    [2, 1]
]
b_eq = [28, 21]

bounds = [(None, None), (None, None)]

res = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method='highs')

print(f"x = {res.x[0]:.2f}, y = {res.x[1]:.2f}")
```
